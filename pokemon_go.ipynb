```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'found_raid' column
df['found_raid'] = label_encoder.fit_transform(df['found_raid'])

# Check value counts
df['found_raid'].value_counts()
```

```python
df.info()
```

```python
df.drop(['pokemon_id'],inplace=True,axis=1)
```

```python
X=df.values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters=i, random_state=5)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,10),wcss)
plt.xlabel("cluster centroids")
plt.ylabel('Wcss')
plt.title('Elbow method')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=5)
kmeans.fit(X)
```

```python
y_kmeans=kmeans.predict(X)
y_kmeans
centroids=kmeans.cluster_centers_
centroids
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], color='g')
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], color='b')
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], color='r')
plt.scatter(centroids[:, 0], centroids[:,1], color='yellow', s=400)
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np


# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=5)
kmeans.fit(X)

# Check the data points in cluster 0
cluster_0_points = X[kmeans.labels_ == 0]
print("Points in cluster 0:", cluster_0_points)

# Check the data points in cluster 1
cluster_1_points = X[kmeans.labels_ == 1]
print("Points in cluster 1:", cluster_1_points)

# Get the indices of points in cluster 0
cluster_0_indices = np.where(kmeans.labels_ == 0)[0]
print("Indices of points in cluster 0:", cluster_0_indices)
```
